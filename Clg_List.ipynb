```python
import fitz  # install PyMuPDF
import re

def extract_text_from_pdf(pdf_path, max_pages=15):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(min(doc.page_count, max_pages)):
        page = doc.load_page(page_num)
        text += page.get_text()
    return text

def clean_text(text):
    # Find all sections between "Score" and "Page"
    sections = re.findall(r'Score(.*?)(Page \d+ of \d+)', text, re.DOTALL)
    cleaned_sections = []

    for section_tuple in sections:
        # Extract the section from the tuple
        section = section_tuple[0]
        section_clean = section.strip()
        cleaned_sections.append(section_clean)

    return "\n\n".join(cleaned_sections)

def remove_repeated_section(text):
    sections_to_remove = [
        "   Minority Seats\n"
        "Seat Type\n"
        "Candidate \n"
        "Category\n"
        "Gender\n"
        "Name of the Candidate\n"
        "Application ID\n"
        "Sr. \n"
        "No.\n"
        "Merit \n"
        "No.\n"
        "Meri\n"
        "Minority Seats Allotted to Minority Candidates\n"
        "MHT-CET \n"
        "Score\n",

        "All India Seats\n"
        "Seat Type\n"
        "Candidate \n"
        "Category\n"
        "Gender\n"
        "Name of the Candidate\n"
        "Application ID\n"
        "Sr. \n"
        "No.\n"
        "Merit \n"
        "No.\n"
        "Meri\n"
        "All India Seats Allotted to All India Candidature Candidates with JEE(Main) Score\n"
        "JEE(Main) \n"
        "Score\n",
        "JEE(Main) \n"
        "Score\n"

    ]

    for section in sections_to_remove:
        text = text.replace(section, "")

    return text

def process_text(text):
    cleaned_text = clean_text(text)
    filtered_text = remove_repeated_section(cleaned_text)
    return filtered_text


pdf_path = "/content/downloaded_file.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
complete_text = process_text(extracted_text)
print(complete_text)
```
